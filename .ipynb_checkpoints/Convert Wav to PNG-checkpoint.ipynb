```python
import matplotlib.pyplot as plt
from matplotlib.backend_bases import RendererBase
from scipy import signal
from scipy.io import wavfile
#import soundfile as sf
import os
import numpy as np
from PIL import Image
from scipy.fftpack import fft

%matplotlib inline
```

```python
#https://www.kaggle.com/timolee/audio-data-conversion-to-images-eda
image_path = './images/'
audio_path = './genres/'
genreList = []
genreNames = []
genres = ['blues', 'classical', 'country', 'disco',
          'hiphop', 'jazz', 'metal', 'pop', 'reggae', 'rock']
#genres = 'blues classical country disco hiphop jazz metal pop reggae rock'.split()

for g in genres:
    if not os.path.exists(image_path + g):
        os.makedirs(image_path + g) 
    if os.path.isdir(audio_path):
        #genreList.append(g + '/')
        for filename in os.listdir(audio_path + g + '/'):
            if os.path.isdir(audio_path + g + '/'):
                genreList.append(g + '/' + filename)   
#print(g)
#print(filename)
#print(genreList)

sample_audio = []
total = 0
for x in genreList:
    all_files = [y for y in os.listdir(audio_path + '/') if '.wav' in y]
    total += len(all_files)
    sample_audio.append(audio_path + x)
    print('count: %d : %s' % (len(all_files), x))
print(total)
#print(sample_audio)
#print(sample_audio)
```

```python
def log_specgram(audio, sample_rate, window_size=20,
                 step_size=10, eps=1e-10):
    nperseg = int(round(window_size * sample_rate / 1e3))
    noverlap = int(round(step_size * sample_rate / 1e3))
    freqs, _, spec = signal.spectrogram(audio,
                                    fs=sample_rate,
                                    window='hann',
                                    nperseg=nperseg,
                                    noverlap=noverlap,
                                    detrend=False)
    return freqs, np.log(spec.T.astype(np.float32) + eps)
```

```python
fig = plt.figure(figsize=(10,10))

# for each of the samples
for i, filepath in enumerate(sample_audio[:9]):
    # Make subplots
    plt.subplot(3,3,i+1)
    
    # pull the labels
    label = filepath.split('/')[-2]
    plt.title(label)
    
    # create spectogram
    samplerate, test_sound  = wavfile.read(filepath)
    _, spectrogram = log_specgram(test_sound, samplerate)
    
    plt.imshow(spectrogram.T, aspect='auto', origin='lower')
    plt.axis('off')
```
